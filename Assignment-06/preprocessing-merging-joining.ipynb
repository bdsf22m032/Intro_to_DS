# Replace all "Inactive" values in the Status column with "Retired".


```python
df['Status'] = df['Status'].replace('Inactive', 'Retired')
df.head()
```

# Add a new column Bonus where values are 10% of the Salary column.


```python
df['Bonus'] = df['Salary'] * 0.10
df.head()
```

# Update the Performance Score of employees in the "IT" department to 5 if it is NaN.


```python
df.loc[(df['Department'] == 'IT') & (df['Performance Score'].isna()), 'Performance Score'] = 5
df.head()
```

# Modify the Experience of employees with less than 2 years to 2.


```python
df.loc[df['Experience'] < 2, 'Experience'] = 2
df.head()
```

# Normalize the Salary column so that all values are between 0 and 1.


```python
df['Normalized_Salary'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())
df.head()
```

# Replace all Location values with "Remote" for employees with a Performance Score below 3.


```python
df.loc[df['Performance Score'] < 3, 'Location'] = 'Remote'
df.head()
```

# Create a new column Experience Level with values "Junior", "Mid", or "Senior" based on the Experience.


```python
def experience_level(experience):
    if experience < 3:
        return 'Junior'
    elif 3 <= experience < 7:
        return 'Mid'
    else:
        return 'Senior'

df['Experience_Level'] = df['Experience'].apply(experience_level)
df.head()
```

```python
df['Age Group'] = df['Age'].apply(lambda x: 'Young' if x < 30 else ('Middle-aged' if x < 50 else 'Senior'))
df.head()
```

```python
import pandas as pd

fp = '/kaggle/input/employee-performance-and-salary-dataset/Employe_Performance_dataset.csv'
df = pd.read_csv(fp)
df.head()
```

# Filter out all rows where the Location is "Chicago".


```python
df = df[df['Location'] == 'Chicago']
df.head()
```

# Delete all rows where the Performance Score is below 2.


```python
df = df[df['Performance Score'] >= 2]
df.head()
```

# Find the average Salary for each Department.


```python
avg_salary_dept = df.groupby('Department')['Salary'].mean()
avg_salary_dept
```

# Calculate the total Salary of all employees grouped by Status.


```python
total_SAL = df.groupby('Status')['Salary'].sum()
total_SAL
```

# Find the maximum Performance Score for employees in each Location.


```python
max_perfo_loc = df.groupby('Location')['Performance Score'].max()
max_perfo_loc
```

# Calculate the total Salary paid to employees who joined in each year.


```python
df['Join_Year'] = pd.to_datetime(df['Joining Date']).dt.year
tS_year = df.groupby('Join_Year')['Salary'].sum()
tS_year
 
```

# Find the department with the highest average Performance Score.


```python
avg = df.groupby('Department')['Performance Score'].mean()
max_perf = avg.idxmax()
max_perf
```

# Count the number of employees in each Location grouped by their Session.


```python
loc_sess = df.groupby(['Location', 'Session']).size()
loc_sess
```

# Group employees by Status and calculate the sum, mean, and standard deviation of their Salary.


```python
sal_stats = df.groupby('Status')['Salary'].agg(['sum', 'mean', 'std'])
sal_stats
```

# Use the apply() method to create a new column Monthly Salary by dividing the Salary by 12.


```python
df['Monthly_Salary'] = df['Salary'].apply(lambda x: x / 12)
df.head()
```

```python
# Use the applymap() method to convert all string values in the DataFrame to uppercase.
```

```python
df = df.applymap(lambda x: x.upper() if isinstance(x, str) else x)
df.head()
```

# Create a second dataset containing Department and Budget. Perform an inner join with the original dataset.


```python
df2 = pd.DataFrame({'Department': ['HR', 'IT', 'Finance'], 'Budget': [100000, 200000, 150000]})
df = pd.merge(df, df2, on='Department', how='inner')
df.head()
```

# Merge the dataset with another dataset that includes additional information about employee ID and their recent Projects.


```python
df3 = pd.DataFrame({'ID': [1, 2, 3], 'Recent_Project': ['Project A', 'Project B', 'Project C']})
df = pd.merge(df, df3, on='ID', how='inner')
df.head()
```

# Perform a left join with a dataset containing Department and Team Lead details, adding Team Lead to the DataFrame.


```python
df4 = pd.DataFrame({'Department': ['HR', 'IT'], 'Team_Lead': ['Alice', 'Bob']})
df = pd.merge(df, df4, on='Department', how='left')
df.head()
```

# Create a DataFrame with employee ID and Project Count, then merge it with the original dataset. Calculate the average Salary per project.


```python
df5 = pd.DataFrame({'ID': [1, 2, 3], 'Project_Count': [5, 3, 4]})
df = pd.merge(df, df5, on='ID', how='inner')
df['Avg_Salary_Per_Project'] = df['Salary'] / df['Project_Count']
df.head()
```

# Calculate the percentage increase in Salary for each employee compared to the average Salary in their Department.


```python
avg_sal = df.groupby('Department')['Salary'].transform('mean')
df['Sal_Increase'] = ((df['Salary'] - avg_sal) / avg_sal) * 100
df.head()
```
