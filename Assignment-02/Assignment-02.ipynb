# Part 2: Immutable Data Management with Tuples

```python
def unique_users_transaction(trans):
    l1 = ()
    duplicates = ()
    unique = ()
    for i in trans:
        if i[1] in l1:
            duplicates+=(i[1],)
        else:
            l1+=(i[1],)

    for i in trans:
        if i[1]  not in duplicates:
            unique+=(i[0],)
    return unique
            



def highest_transaction(trans):
    high = trans[2]
    for i in trans:
        if high[2]<i[2]:
            high = i
    return high




def separate_list(trans1):
    l1 =[]
    l2 =[]
    for i in trans1:
        l1.append(i[0])
        l2.append(i[1])
    return l1,l2


    

trans = (
    (1, 101, 250.75, "2024-10-14 09:30:00"),
    (2, 102, 150.00, "2024-10-13 09:30:00"),
    (3, 101, 250.75, "2024-10-12 09:30:00"),
    (4, 103, 300.50, "2024-10-11 09:30:00"),
    (5, 104, 500.00, "2024-10-10 09:30:00"),
    (6, 102, 150.00, "2024-10-09 09:30:00"),
    (7, 105, 4500.25, "2024-10-08 09:30:00"),
    (8, 104, 75.50, "2024-10-07 09:30:00"),
    (9, 101, 250.75, "2024-10-06 09:30:00"),
    (10, 107, 600.00, "2024-10-05 09:30:00")
)

trans1 = [
    (1, 101, 250.75, "2024-10-14 09:30:00"),
    (2, 102, 150.00, "2024-10-13 09:30:00"),
    (3, 101, 250.75, "2024-10-12 09:30:00"),
    (4, 103, 300.50, "2024-10-11 09:30:00"),
    (5, 104, 500.00, "2024-10-10 09:30:00"),
    (6, 102, 150.00, "2024-10-09 09:30:00"),
    (7, 105, 4500.25, "2024-10-08 09:30:00"),
    (8, 104, 75.50, "2024-10-07 09:30:00"),
    (9, 101, 250.75, "2024-10-06 09:30:00"),
    (10, 107, 600.00, "2024-10-05 09:30:00")
]



print('Unique Usersids : ')
x =unique_users_transaction(trans)
print(x)

print('----------------------------------------------------------------------------------------')
print('Highest Transactions')
y =highest_transaction(trans)
print(y)



print('----------------------------------------------------------------------------------------')
print("Separate lists for Transactionsids  and Userids :")
z =separate_list(trans1)
l1,l2 =z
print('Transcations ids:',l1)
print('User ids : ',l2 )
```
